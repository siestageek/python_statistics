# 확률 
* 오랜시간동안 경험적으로 쌓아온 확률현상에 대한 경험적 인식을 바탕으로 이론화한 것.
+ 어떤 일이 발생할 가능성이나 믿음의 정도를 나타낸 것
+ $0$과 $1$사이의 숫자를 그 사건 $A$가 일어날 확률
+ 어떤 일이 일어날 경우의 수 / 모든 일이 일어날 경우의 수
+ $1$ - 일어날 확률 = 일어나지 않을 확률

# 확률의 종류
1. 수학적 확률 
  + 어떤 사람이 계산하던지간에 동일한 값으로 계산되는 확률 (고전적/경험적 확률)
  + 동전 던지기, 주사위 굴리기, 카드 뽑기 등 표본공간의 결과(표본점)가 유한하고, 
  + 모든 결과가 발생할 가능성이 같을때

2. 통계적 확률 
  + 동일조건&독립적으로 무한반복했을 때 발생하는 확률
  + 어떤 시행을 반복할 때, $N$번 시행에서 사건 $A$가 발생한 횟수를 $n(A)$라 할 때
  + 이 시행을 무한히 반복한다고 할 때, 즉 상대도수의 극한으로 확률을 정의하는 방법

3. 주관적 확률 
  + 관찰자의 주관에 따라 다르게 표현되는 확률  
  + 한 개인의 경험이나 지식, 정보, 직관 등을 토대로 각자의 주관에 따라 확률을 정함
  + 특히 내기와 도박에서 많이 사용

# 표본공간 
 * "통계적 실험에서 발생가능한 모든 결과들의 집합"으로 정의하고 이를 $S$로 표기
    + 두 개의 동전을 던지는 실험에서 앞면과 뒷면이 나오는 표본공간은?
       - 앞앞 앞뒤 뒤앞 뒤뒤

    + 세 개의 동전을 던지는 실험에서의 앞면과 뒷면이 나오는 표본공간은?


  * 표본공간은 모든 결과의 집합을 나타내기 때문에, 이 결과들을 가지고 부분집합을 만들 수 있음

# 확률적 사건  
  * 표본 공간에서 특정한 조건을 만족하는 결과를 모아 놓은 집합
  * 간단하게 사건(event)이라 함
  * 여기서 알 수 있는 것은 사건은 표본공간 밖에서 만들어질 수 없음




# 확률적 사건의 종류

표본공간 상에서 어떤 부분집합이냐에 따라 사건의 이름이 달라짐

1. 전사건(total event) : 표본공간 $S$의 모든 원소를 포함하는 사건.
2. 공사건(null event) : 표본공간 $S$의 어떤 원소도 포함하지 않는 사건.
3. 여사건(complementary event) : 포본공간 $S$의 사건 E에 속하지 않는 $S$의 모든 원소들의 집합인 사건
4. 합사건(union event) : 두 사건 $E$와 $F$에 대하여 $E$ 또는 $F$ 중 적어도 한 쪽은 일어나는 사건
5. 곱사건(intersection event) : 두 사건 $E$와 $F$에 대하여 $E$와 $F$가 동시에 일어나는 사건
6. 배반사건(mutually exclusive event) : 두 사건 $E$와 $F$에 대하여 $E$와 $F$중 한 쪽이 일어나면 다른 쪽은 일어나지 않을 때,    $E$와 $F$는 서로 배반이라 한다.(서로 공통부분이 없다.)


# 가장 일반적이고 전통적인 확률의 정의
 * 모든 사건의 경우의 수'에 대한 '특정한 사건이 발생한 빈도수'의 비율
 * 결국 확률은 '비율(ratio)' = 할당된 확률
 * 사건 A의 각각의 원소에 가중치(weight)가 부여됐다는 의미



# 확률의 덧셈과 곱셈 : 
* 두 사건이 동시에 일어나면 확률을 곱하고, 동시에 일어나지 않으면 더함

* 사건 $A$ 또는 사건 $B$가 일어날 확률   
    = (사건 $A$가 일어날 확률) + (사건 $B$가 일어날 확률)

* 사건 $A$와 사건 $B$가 동시에 일어날 확률   
    = (사건 $A$가 일어날 확률) × (사건 $B$가 일어날 확률)

* **두 사건 $A$, $B$에 대하여**
  + **합사건**
      - $A$ 또는$B$가 일어나는 사건 : $A \cup B$
  + **곱사건**
      - $A$와 $B$가 동시에 일어나는 사건 : $A \cup B$

# 조건부 확률(conditional probability) 
 * 베이즈 정리(Bayes' Theorem)의 시작
 * 한 사건이 일어났다는 전제 하에서 다른 사건이 일어날 확률
 * 어떤 사건이 일어날 가능성에 대한 믿음을 더욱 확고하게 하려면, 그 사건에 대한 많은 정보가 필요할 수 있음
 * 즉, 바로 주어진 정보에 따라 확률이 달라질 수 있다는 의미
 * 조건부확률에서 핵심은 주어진 정보로 인해서 전체 표본공간을 고려할 필요가 없다는 것
 * 즉 주어진 정보로 인해 표본공간이 사건 B로 축소되었다고 해석
 * 한 사건의 발생이 다른 사건의 발생 확률과 상관관계가 있음을 의미

 $$ P(A|B) = { P(A \cup B) \over P(B) } $$

 >   ex) 한국여자 X, 한국남자 Y, 미국여자 Z가 어떤 회사에 취업면접을 봤음    
 세사람 중 반드시 한 명은 취업이 된다고 가정하면, 한국여자 X가 취업될 확률은 1/3 정도이지만,   
회사 사정에 의해, 여성 지원자만 선출한다는 사실이 공개되는 경우,   
    한국여자 X가 취업될 확률은 1/2 로 상승했다고 보는 것이 합리적

 > 전체 국민 중 군 복무자의 비율이 40%라 하자. 이 중, 남성이 80%라 할 때,   
전체 남성 중 무작위로 선정한 한 사람이 군 복무자일 확률은? (전 국민의 성비는 5:5라 한다.)
  
* $$ P(군복무자 | 남성) = 
{ P(군복무자 \cup 남성) \over P(남성) }
= 0.32 = 64% $$

# 전체 확률의 법칙(Law of Total)
 * 몇 가지 별개의 사건을 통해 실현될 수 있는 결과의 전체 확률
 * 특정 사건의 조건부 확률은 개별 조건부 확률의 합: 조건부 확률들의 덧셈
 * 베이즈 정리(Bayes' Theorem)의 중요 개념
 * 사건 $A$ 의 확률은 사건 $A$ 와 복수의 사건 $C_i$ 가 동시에 발생할 사건들의 확률의 합과 같다는 법칙
   $$ P(A) = \sum _i P(A \cup C_i) $$

> 어떤 나라에서 시민의 39%가 여성이라고 가정해 보자.    
우리는 또한 여성의 22%와 남성의 14%가 직업이 없다는 것을 알고 있다.   
그렇다면 이 나라의 인구 중에서 무작위로 뽑은 사람이 실업자 P(U)일 확률은 얼마나 될까?

* 실업자일 확률은 뽑은사람이 남성이면서 실업자일 확률과 
여성이면서 실업자일 확률의 총합
* $ P(U) = P (W) P (U | W) + P (M) P(U | M) $   
 
* 그 사람이 여성일 확률 : $P(W) = 0.39 $
* 그 사람이 남성일 확률 : $P(M) = 1 - 0.39 = 0.61 $

* 또한 명시된 문제는 조건부 확률을 제공한다.
  - 여성이 실업자일 확률 : $ P(U|W) = 0.22 $
  - 남성이 실업자일 확률 : $ P(U|M) = 0.14 $

* 따라서 총 확률 법칙을 사용하면 다음과 같은 결과를 얻을 수 있다.
  - $ P (U) $ = 0.22 × 0.39 + 0.14 × 0.61 = 0.17 


# 베이즈 정리(bayes' theorem, bayes' rule)
* 이전의 경험과 현재의 증거를 토대로 어떤 사건의 확률을 추론
* 확률 P(A|B)를 알고 있을 때, 관계가 정반대인 확률 P(B|A)를 계산하기 위해 등장
* 즉, '결과를 관측한 뒤 원인을 추론할 수 있다'는 것

  + $P(A)$ : $A$의 사전 확률(evidence) - 현재의 증거    
  + $P(B)$ : $B$의 사전 확률(prior probability) - 과거의 경험
  + $P(A | B)$ : 사건 $B$가 주어졌을 때 $A$의 조건부 확률(likelihood) 
     - 알려진 결과(관찰된 결과)에 기초한 어떤 가설에 대한 가능성(즉, 이 가설을 지지하는 정도)
  + $P(B | A)$ : 사건 $A$라는 증거에 대한 사후 확률(posterior probability) 
     - 사건 $A$가 일어났다는 것을 알고, 그것이 사건 $B$로부터 일어난 것이라고 생각되는 조건부 확률

# 확률변수
* 동전을 2개 던질 때 앞면이 나오는 횟수를 $X$ 라고 할때
* 표본공간의 각 원소 {앞,앞}, {앞,뒤}, {뒤,앞}, {뒤,뒤} 각각에 하나의 실수값(2,1,0)을 대응됨
* 표본공간의 각 원소에 하나의 실수를 대응시킬 때 이 실수를 '확률변수(random variable)'라고 함
  + 따라서 확률변수는 함수를 의미
* 확률변수 $X$의 집합을 셀 수 있으면 '이산확률변수', 
* 확률변수 $X$가 연속적인 구간 내의 값을 취하면 '연속확률변수'라 함
* 이산확률변수와 연속확률변수의 가장 큰 차이점은 $P(X=x)$로 표현할 수 있는지 없는지임

 현실적으로 확률이 0에 가까움 - 정확히 100ml 따르는 사건이 발생할 수도 있음

# 확률질량함수
* 이산확률변수의 확률분포를 나타내는 함수
* 이산 확률 변수에서 특정 값에 대한 확률을 나타내는 함수
* 확률변수가 취할 수 있는 값이 유한개이거나 자연수와 같이 셀 수 있는 이산확률변수일 때,    
그 불연속한 값에 대한 확률을 나타내는 함수

* 동전을 3개 던질 때 앞면이 나오는 횟수를 $X$ 라고 할때 확률변수는 3,2,1,0 임
* 각 확률변수 $X$ 에 관한 함수를 만들어 보면  0=1/8, 1=3/8, 2=3/8, 3=1/8
* 확률질량함수는 말 그대로 '함수'이므로 그래프로 나타낼 수 있음

# 확률밀도함수
* 연속확률변수의 확률분포를 나타내는 함수
* 확률변수가 취할 수 있는 값이 연속적이며 무한하기 때문에 분포를 표현하는 것이 불가능
* 따라서, 확률변수 $X=?$일 때 확률은 현실적으로 0에 가까움
  + 물을 정확히 100.00ml 따를 수 있는 확률은 0에 가까움
  + 반면, 95ml ~ 100ml 따를 수 있는 확률은 적어도 0은 아님
* 따라서, 연속확률변수의 확률을 구할 때는 항상 범위로 표현!
* 연속확률변수에서 부등호에 붙은 등호는 의미가 없음
* 특정 범위의 확률을 구하는 것은 확률밀도함수 그래프의 넓이를 구하는 것 - 적분 이용
